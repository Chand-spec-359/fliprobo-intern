```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
driver = webdriver.Chrome(r"F:\AAAAAAAA\chromedriver_win32\chromedriver.exe")
```

```python
# Opening the archive page
driver.get('https://www.thehindu.com/archive/web/2016/03/')
```

```python
#Dates Urls
date_urls = []
date = driver.find_elements_by_xpath('//a[@class="ui-state-default"]')
for i in date:
    date_urls.append(i.get_attribute('href'))

date_urls
```

```python
## first date news URLs
news_url = []
for date in date_urls[0:4]:  # scrape news url from 4 dates
    driver.get(date)
    news = driver.find_elements_by_xpath(("//ul[@class='archive-list']/li/a"))
    for i in news:
        news_url.append(i.get_attribute("href"))
news_url
```

```python
len(news_url)
```

```python
# taking 10 news_url to scrape headline
headline = []
for url in news_url[0:10]:
    driver.get(url)
    head = driver.find_elements_by_xpath(('//h1[@class="title"]'))
    for i in head:
        headline.append(i.text)
        
headline
```
