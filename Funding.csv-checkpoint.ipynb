```python
import selenium
import pandas as pd
import time
from selenium import webdriver
```

```python
driver= webdriver.Chrome('chromedriver.exe')
```

```python
url='https://trak.in/'
driver.get(url)
```

```python
search_bar= driver.find_element_by_id('menu-item-51510') #locating search_bar by id
search_bar.click()
```

```python
serial_tags=driver.find_elements_by_xpath("//td[@class='column-1']")
serial_tags
serial_list=[]
for i in serial_tags:
    serial_list.append(i.text)
serial_list
```

```python
Date_tags=driver.find_elements_by_xpath("//td[@class='column-2']")
Date_tags
Date_list=[]
for i in Date_tags:
    Date_list.append(i.text)
Date_list
```

```python
startup_tags=driver.find_elements_by_xpath("//td[@class='column-3']/a")
startup_tags
startup_list=[]
for i in startup_tags:
    startup_list.append(i.text)
startup_list
```

```python
Industry_tags=driver.find_elements_by_xpath("//td[@class='column-4']")
Industry_tags
Industry_list=[]
for i in Industry_tags:
    Industry_list.append(i.text)
Industry_list
```

```python
Subvertical_tags=driver.find_elements_by_xpath("//td[@class='column-5']")
Subvertical_tags
Subvertical_list=[]
for i in Subvertical_tags:
    Subvertical_list.append(i.text)
Subvertical_list
```

```python
city_tags=driver.find_elements_by_xpath("//td[@class='column-6']")
city_tags
city_list=[]
for i in city_tags:
    city_list.append(i.text)
city_list
```

```python
investor_tags=driver.find_elements_by_xpath("//td[@class='column-7']")
investor_tags
investor_list=[]
for i in investor_tags:
    investor_list.append(i.text)
investor_list
```

```python
investment_tags=driver.find_elements_by_xpath("//td[@class='column-8']")
investment_tags
investment_list=[]
for i in investment_tags:
    investment_list.append(i.text)
investment_list
```

```python
amount_tags=driver.find_elements_by_xpath("//td[@class='column-9']")
amount_tags
amount_list=[]
for i in amount_tags:
    amount_list.append(i.text)
amount_list
```

# JULY'2020

```python
Funding=pd.DataFrame({})
Funding['Serial No.']=serial_list[75:85]
Funding['Date']=Date_list[75:85]
Funding['Startup Name']=startup_list[75:85]
Funding['Industry/Vertical']=Industry_list[75:85]
Funding['Sub-Vertical']=Subvertical_list[75:85]
Funding['City/Location']=city_list[75:85]
Funding['Investor Name']=investor_list[75:85]
Funding['Investment Type']=investment_list[75:85]
Funding['Amount(In USD)']=amount_list[75:85]
Funding
```
