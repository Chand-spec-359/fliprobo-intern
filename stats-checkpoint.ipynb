```python
#import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "http://statisticstimes.com/"
driver.get(url)
time.sleep(6)
```

```python
Economy=driver.find_element_by_xpath("//*[@id='top']/div[2]/div[2]/button")
Economy.click()
```

```python
try:
    India_dropdown=driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/div/a[3]')
    
except ElementNotInteractableException as e:
    print("Exception Raised: ", e)
```

```python
GDP=driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/ul/li[1]/a")
GDP.click()
```

```python
Rank=[]
State=[]
GSDP1=[]
GSDP2=[]
Share=[]
GDP=[]

rank=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[1]")
for i in rank:
    Rank.append(i.text)

state=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[2]")
for i in state:
    State.append(i.text)
    
gsdp1=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[3]")
for i in gsdp1:
    GSDP1.append(i.text)
    
gsdp2=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]")
for i in gsdp2:
    GSDP2.append(i.text)
    
share=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[5]")
for i in share:
    Share.append(i.text)
    
gdp=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]")
for i in gdp:
    GDP.append(i.text)
```

```python
df=pd.DataFrame({"Rank":Rank,"State":State,"GSDP at current price (19-20)":GSDP1
                ,"GSDP at current price (18-19)":GSDP2,"Share(18-19)":Share,"GDP($ billion)":GDP})
df
```
