```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
driver= webdriver.Chrome('chromedriver.exe')
```

```python
url='https://www.thehindu.com/archive/web/2020/11/'
driver.get(url)
```

```python
#Dates Urls
date_urls = []
date = driver.find_elements_by_xpath('//a[@class="ui-state-default"]')
for i in date:
    date_urls.append(i.get_attribute('href'))

date_urls
```

```python
## news URLs
news_url = []
for date in date_urls[0:30]:  # scrape news url from 30 dates
    driver.get(date)
    news = driver.find_elements_by_xpath(("//ul[@class='archive-list']/li/a"))
    for i in news:
        news_url.append(i.get_attribute("href"))
news_url
```

```python
# length of news URL's
len(news_url)
```

```python
# Headlines for all 30 dates extracted
headline = []
for url in news_url[0:500]:
    driver.get(url)
    head = driver.find_elements_by_xpath(('//h1[@class="title"]'))
    for i in head:
        headline.append(i.text)
        
headline
```

```python
#Description for first 1000 news extracted

description = []
for news in news_url[0:500]:
    driver.get(news)
    des = driver.find_elements_by_xpath('/html/body/div[2]/div[2]/div/div/div/section/div/div/div/div[3]/div[3]')
    for i in des:
        description.append(i.text)
description
```

```python
# Dates for first 1000 news extracted
dates = []

for news in news_url[0:500]:
    driver.get(news)
    date = driver.find_elements_by_xpath("//span[@class='blue-color ksl-time-stamp']/none")
    for i in date:
        dates.append(i.text)  
    
dates
```

```python
# Author for first 1000 news extracted

Author = []

for news in news_url[0:1000]:
    driver.get(news)
    Auth = driver.find_elements_by_xpath("//a[@class='auth-nm lnk']")
    for i in Auth:
        Author.append(i.text)  
    
Author 
    
```

```python
# State categories for first 1000 dates extracted

Category = []

for news in news_url[0:500]:
    driver.get(news)
    Auth = driver.find_elements_by_xpath("//a[@class='section-name']")
    for i in Auth:
        Category.append(i.text)  
    
Category
```
