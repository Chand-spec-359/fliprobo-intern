```python
import selenium
import pandas as pd
import time
from selenium import webdriver
```

```python
driver= webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.amazon.in/')
user_inp=input('enter the product you want to search :')
```

```python
# /span is child tag of div

search_bar= driver.find_element_by_id('twotabsearchtextbox') #locating search_bar by id
search_bar.clear() # clearing search_bar if something is written
search_bar.send_keys(user_inp) # sending user_inp to search bar
time.sleep(4)
search_btn= driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']/span/input")
search_btn.click()
```

```python
#fetching URL's
urls=[]
for i in range(0,3):
    for j in driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]/a'):
        urls.append(j.get_attribute('href'))
    driver.find_element_by_xpath('//li[@class="a-last"]/a').click()
    time.sleep(4)
```

```python
urls
```

```python
name=[]
image_url=[]
price=[]
```

```python
for i in range(0,3):
    for j in driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]'):
        name.append(j.text)
    for k in driver.find_elements_by_xpath('//span[@class="a-price-whole"]'):
        price.append(k.text)
    for m in driver.find_elements_by_xpath('//div[@class="a-section aok-relative s-image-fixed-height"]/img'):
        image_url.append(m.get_attribute("src"))
    driver.find_element_by_xpath('//li[@class="a-last"]/a').click()
    time.sleep(4)
```
